```python
import random

def tic_tac_toe():
    # Initialize the board
    board = [' ' for _ in range(9)]
    player_icon = ''
    computer_icon = ''
    current_turn = 'Player' if random.randint(0, 1) == 0 else 'Computer'

    def print_board():
        print(f'''
  {board[0]} | {board[1]} | {board[2]}
 -----------
  {board[3]} | {board[4]} | {board[5]}
 -----------
  {board[6]} | {board[7]} | {board[8]}
        ''')

    def choose_icon():
        nonlocal player_icon, computer_icon
        icon = input("Choose your icon (X or O): ").upper()
        while icon not in ['X', 'O']:
            print("Invalid choice. Please choose X or O.")
            icon = input("Choose your icon (X or O): ").upper()
        player_icon = icon
        computer_icon = 'O' if player_icon == 'X' else 'X'

    def check_win(icon):
        win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0, 3, 6), (1, 4, 7), (2, 5, 8),
                          (0, 4, 8), (2, 4, 6)]
        for condition in win_conditions:
            if board[condition[0]] == board[condition[1]] == board[condition[2]] == icon:
                return True
        return False

    def check_draw():
        return ' ' not in board

    def player_move():
        try:
            move = int(input("Choose your move (1-9): ")) - 1
            if board[move] == ' ':
                board[move] = player_icon
            else:
                print("This space is already taken.")
                player_move()
        except (ValueError, IndexError):
            print("Invalid move. Please choose a number between 1 and 9.")
            player_move()

    def computer_move():
        available_spaces = [i for i, x in enumerate(board) if x == ' ']
        board[random.choice(available_spaces)] = computer_icon

    def game_loop():
        nonlocal current_turn
        print(f"{current_turn} will go first.")
        while True:
            print_board()
            if current_turn == 'Player':
                player_move()
                if check_win(player_icon):
                    print_board()
                    print("Congratulations! You have won!")
                    break
                current_turn = 'Computer'
            else:
                computer_move()
                if check_win(computer_icon):
                    print_board()
                    print("Computer wins!")
                    break
                current_turn = 'Player'
            if check_draw():
                print_board()
                print("It's a draw!")
                break

    choose_icon()
    game_loop()

if __name__ == "__main__":
    tic_tac_toe()
```
